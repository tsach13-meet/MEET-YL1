```python
import csv
from pylab import *
f = open('environmental.csv', 'rb')
reader = csv.reader(f)
rownum = 0
x=[]
y=[]
for row in reader:
    
    if rownum != 0:
        a = row[0:2]
        a = map(float, a)
        x.append(a[0])
        y.append(a[1])
        
    rownum+= 1
  
 
f.close()
      #savefig      
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(x,y)
ax.set_title('graf  radiation as fanction of ozone')
ax.set_xlabel('ozone()')
ax.set_ylabel('radiation(m/s)')
      
```
