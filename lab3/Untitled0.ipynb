```python
import csv
from pylab import *
f = open('cig.csv', 'rb')
reader = csv.reader(f)
rownum = 0
x=[]
y=[]


for row in reader:
    s= row[0]
    s= s.replace("\t", " ")
    s= s.split()
    s= s[1:]
    s= map(float,s)
    #print s
    x.append(s[0])
    y.append(s[1])
    rownum += 1
print x
print y 


fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(x,y)
ax.set_title('graf speed as fanction of time')
ax.set_xlabel('t(sec)')
ax.set_ylabel('v(m/s)')

f.close()
savefig("cig_scatter.png")

```
